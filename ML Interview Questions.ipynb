# ML Interview Questions

A set of interview questions I can think of and explanations or references with the explanations or without either.

1) Given matrix of data points $X$ (sometimes called design matrix - with first rows being all ones) which is $n$ (number of samples) by $m+1$ (bias plus number of features) and a vector of target values $\vec{y}$ of length $n$. Find the weights $\vec{w}$ (size m+1) that minimizes the mean-squared error loss function: $|| X\vec{w} - \vec{y}||^2$. Can you give an analytic expression when $L_2$ regularization is used? What if you use $L_1$?

    - Answer: See "The Elements of Statistical Learning". You can do it doing simple calculus on vector by differentiating loss function and setting the derivative to zero. The analytic expressions can be given in terms of plain and ridge regression.

2) Explain the different behaviour of $L_2$ (Ridge regression) and $L_1$ (Lasso regression) regularizations. 

    - Answer: Lasso regression often leads to some of the weights being 0. This can be somehow explained by looking at the gradients which are in this case constant and can lead to coefficients being zero whereas in case of the Rigde regression the gradients are proportional to the values and can only assymptotically lead to them being zero. Another way to see this is to look at this visualization from "The Elements of Statistical Learning". The visualizations depicts how it the countour (of constant loss function) is more likely to cross an axis than to be tangent to a circle (hyper sphere).
